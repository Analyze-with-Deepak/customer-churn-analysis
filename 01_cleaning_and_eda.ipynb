```python
pd.crosstab(df["Contract"], df["Churn"])
```

```python
contract_churn_rate = pd.crosstab(df["Contract"], df["Churn"], normalize="index") * 100
contract_churn_rate.round(2)
```

```python
contract_churn_rate["Yes"].plot(kind="bar")
plt.title("Churn Rate by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Churn Rate (%)")
plt.show()
```

Customers with month-to-month contracts showed the highest churn rate, while customers with one-year and two-year contracts had lower churn. This suggests that longer contract commitments may be associated with stronger customer retention.

```python
contract_churn_rate["Yes"].sort_values(ascending=False)
```

Customers who churned had lower average tenure than retained customers, suggesting that newer customers may be at higher risk of leaving. This indicates that early customer experience and onboarding may be important for retention.

```python
df["ChurnFlag"] = df["Churn"].map({"No": 0, "Yes": 1})
```

```python
high_risk_segments = (
    df.groupby(["Contract", "PaymentMethod"])
      .agg(
          customers=("customerID", "count"),
          churn_rate=("ChurnFlag", "mean"),
          avg_monthly_charges=("MonthlyCharges", "mean"),
          avg_tenure=("tenure", "mean")
      )
      .reset_index()
)
```

```python
high_risk_segments["churn_rate"] = (high_risk_segments["churn_rate"] * 100).round(2)
high_risk_segments["avg_monthly_charges"] = high_risk_segments["avg_monthly_charges"].round(2)
high_risk_segments["avg_tenure"] = high_risk_segments["avg_tenure"].round(2)

high_risk_segments = high_risk_segments.sort_values(
    by=["churn_rate", "customers"],
    ascending=[False, False]
)

high_risk_segments.head(10)
```

```python
top_segments = high_risk_segments.head(5).copy()
top_segments["Segment"] = top_segments["Contract"] + " | " + top_segments["PaymentMethod"]
top_segments[["Segment", "customers", "churn_rate", "avg_monthly_charges", "avg_tenure"]]
```

```python
top_segments.plot(kind="bar", x="Segment", y="churn_rate")
plt.title("Top 5 High-Risk Customer Segments")
plt.xlabel("Customer Segment")
plt.ylabel("Churn Rate (%)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

Segment-level analysis showed that some customer groups had much higher churn rates than others. Combining contract type and payment method helped identify higher-risk segments more clearly than looking at a single variable alone.

```python
findings_summary = pd.DataFrame({
    "Finding": [
        "Month-to-month contracts had higher churn",
        "Certain payment methods had higher churn",
        "Churned customers had different monthly charges",
        "Churned customers had shorter tenure"
    ],
    "Business Meaning": [
        "Short-term commitment customers may be less loyal",
        "Payment behavior may signal higher churn risk",
        "Pricing may influence retention",
        "Newer customers may be more likely to leave"
    ]
})

findings_summary
```

```python
recommendations = pd.DataFrame({
    "Recommendation": [
        "Target month-to-month customers with retention offers",
        "Review the customer experience for higher-risk payment groups",
        "Improve onboarding for newer customers"
    ],
    "Reason": [
        "This contract group showed higher churn risk",
        "Some payment segments appeared more likely to churn",
        "Customers with shorter tenure appeared more likely to leave"
    ]
})

recommendations
```

```python
findings_summary.to_csv("findings_summary.csv", index=False)
recommendations.to_csv("business_recommendations.csv", index=False)
```

Based on the analysis, I recommended focusing on higher-risk customer groups such as month-to-month customers, investigating payment-method-related churn patterns, and improving retention efforts for newer customers. These recommendations were based on segment-level churn patterns and key customer characteristics.
